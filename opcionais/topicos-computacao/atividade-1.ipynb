```python
import pandas as pd
from sklearn.datasets import load_wine

# Carregando o conjunto de dados Wine
wine = load_wine()

# Convertendo os dados para um DataFrame do pandas
df = pd.DataFrame(data=wine.data, columns=wine.feature_names)

# Calculando a média de cada coluna
media = df.mean()

# Calculando a mediana de cada coluna
mediana = df.median()

# Calculando o desvio padrão de cada coluna
desvio_padrao = df.std()

# Calculando a variância de cada coluna
variancia = df.var()

# Exibindo os resultados
resultados = pd.DataFrame({
    'Média': media,
    'Mediana': mediana,
    'Desvio Padrão': desvio_padrao,
    'Variância': variancia
})
print(resultados)
```

```python
import matplotlib.pyplot as plt

# Loop sobre cada atributo para criar e exibir os gráficos
for atributo in resultados.index:
    # Box plot
    plt.figure(figsize=(8, 6))
    plt.boxplot(resultados.loc[atributo])
    plt.title(f'Box Plot de {atributo}')
    plt.ylabel('Valor')
    plt.grid(True)
    plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine

# Carregar o conjunto de dados Wine
wine = load_wine()
data = wine.data
feature_names = wine.feature_names

# Escolher a variável que você deseja visualizar no histograma
# Neste exemplo, vamos escolher a primeira característica (feature)
feature_index = 0
feature_name = feature_names[feature_index]
feature_data = data[:, feature_index]

# Criar o histograma
plt.hist(feature_data, bins=20, color='blue', alpha=0.7)
plt.xlabel(feature_name)
plt.ylabel('Frequência')
plt.title('Histograma de {}'.format(feature_name))
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine

# Carregar o conjunto de dados Wine
wine = load_wine()
data = wine.data
feature_names = wine.feature_names

# Definir o número de bins para o histograma
num_bins = 20

# Criar subplots para os histogramas
fig, axes = plt.subplots(4, 3, figsize=(15, 15))
fig.suptitle('Histogramas dos Atributos do Conjunto de Dados Wine')

# Iterar sobre todas as características e criar um histograma para cada uma delas
for i, ax in enumerate(axes.flat):
    if i < len(feature_names):
        feature_data = data[:, i]
        ax.hist(feature_data, bins=num_bins, color='blue', alpha=0.7)
        ax.set_xlabel('Valor')
        ax.set_ylabel('Frequência')
        ax.set_title(feature_names[i])

# Ajustar o layout
plt.tight_layout()
plt.show()
```
